```python
import bs4
from bs4 import BeautifulSoup as bs
import requests
```

```python
link='https://www.flipkart.com/search?q=mobile+phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&p%5B%5D=facets.sim_type%255B%255D%3DSingle%2BSim&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D25000&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG'
```

```python
page = requests.get(link)
```

```python
page
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

# Univariate Analysis Plot Histograms and BoxPlots

```python
# Plot Histograms of Price and Rating
import seaborn as sns
import matplotlib.pyplot as plt

df1.hist(figsize=(14,5))
plt.show()
```

```python
# Plot Distibution Plots of Price and Rating
columns=['Price','Rating']
for i in columns:
    sns.kdeplot(df1[i],shade=True)
    plt.xlabel(i, fontsize=18)
    plt.ylabel('Rating', fontsize=16)
    plt.show()
```

```python
# Boxplot of Price  using Dataframe method
df1.boxplot(column='Price',grid=True,figsize=(6,4))
plt.show()
```

```python
# Box plot of Rating

df1.boxplot(column='Rating',grid=True,figsize=(6,4))
plt.show()
```

```python
# Bivariate Analysis
# Box plot of CPU and Price
plt.figure(figsize=(15,8))
sns.boxplot(y="Rating",x='Price',data=df1)
plt.show()
```
